```python
# Concrete Compressive Strength Prediction using Keras
## Part A: Build a Baseline Model

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam

# Load data
url = "https://cocl.us/concrete_data"
data = pd.read_csv(url)

# Features and target variable
X = data.drop('Strength', axis=1)
y = data['Strength']

# Initialize lists to store mean squared errors
mse_list = []

# Repeat 50 times
for _ in range(50):
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=np.random.randint(1000))
    
    # Define the model
    model = Sequential()
    model.add(Dense(10, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(1))
    
    # Compile the model
    model.compile(optimizer=Adam(learning_rate=0.01), loss='mean_squared_error')
    
    # Train the model
    model.fit(X_train, y_train, epochs=50, verbose=0)
    
    # Predict and evaluate the model
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_list.append(mse)

# Compute mean and standard deviation of the MSEs
mse_mean = np.mean(mse_list)
mse_std = np.std(mse_list)

print(f"Mean Squared Error Mean: {mse_mean:.4f}")
print(f"Mean Squared Error Standard Deviation: {mse_std:.4f}")
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize lists to store mean squared errors
mse_list_normalized = []

# Repeat 50 times
for _ in range(50):
    # Normalize features
    scaler = StandardScaler()
    X_normalized = scaler.fit_transform(X)
    
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.3, random_state=np.random.randint(1000))
    
    # Define the model
    model = Sequential()
    model.add(Dense(10, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(1))
    
    # Compile the model
    model.compile(optimizer=Adam(learning_rate=0.01), loss='mean_squared_error')
    
    # Train the model
    model.fit(X_train, y_train, epochs=50, verbose=0)
    
    # Predict and evaluate the model
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_list_normalized.append(mse)

# Compute mean and standard deviation of the MSEs
mse_mean_normalized = np.mean(mse_list_normalized)
mse_std_normalized = np.std(mse_list_normalized)

print(f"Mean Squared Error Mean (Normalized): {mse_mean_normalized:.4f}")
print(f"Mean Squared Error Standard Deviation (Normalized): {mse_std_normalized:.4f}")
```

```python
# Initialize lists to store mean squared errors
mse_list_layers = []

# Repeat 50 times
for _ in range(50):
    # Normalize features
    scaler = StandardScaler()
    X_normalized = scaler.fit_transform(X)
    
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.3, random_state=np.random.randint(1000))
    
    # Define the model with three hidden layers
    model = Sequential()
    model.add(Dense(10, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    
    # Compile the model
    model.compile(optimizer=Adam(learning_rate=0.01), loss='mean_squared_error')
    
    # Train the model with 50 epochs
    model.fit(X_train, y_train, epochs=50, verbose=0)
    
    # Predict and evaluate the model
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_list_layers.append(mse)

# Compute mean and standard deviation of the MSEs
mse_mean_layers = np.mean(mse_list_layers)
mse_std_layers = np.std(mse_list_layers)

print(f"Mean Squared Error Mean (Three Hidden Layers): {mse_mean_layers:.4f}")
print(f"Mean Squared Error Standard Deviation (Three Hidden Layers): {mse_std_layers:.4f}")
```
